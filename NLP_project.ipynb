```python
data=pd.read_excel('Sentence.xlsx','Sheet1')
```

```python
data.head()
```

```python
ref_sent=data.iloc[4][1]
ref_sent
```

```python
ref_sent_vec=nlp(ref_sent)
ref_sent_vec
```

```python
type(ref_sent_vec)
```

```python
data.shape
```

```python
all_docs=[nlp(row) for row in data['Message']]
```

```python
all_docs
```

```python
len(all_docs)
```

```python
type(all_docs[0])
```

```python
all_docs[0]
```

```python
sims=[]
doc_id=[]
for i in range(len(all_docs)):
    sim=all_docs[i].similarity(ref_sent_vec)
    sims.append(sim)
    doc_id.append(i)
```

```python
sims_doc=pd.DataFrame(list(zip(doc_id,sims)),columns=("doc_id","sims"))
```

```python
sims_doc.head(10)

```

```python
sims_doc_sorted=sims_doc.sort_values(by='sims',ascending=False)
```

```python
sims_doc_sorted.head()
```

```python
sims_doc_sorted['doc_id'][0:5]
```

```python
top5_sim_doc = data.iloc[sims_doc_sorted['doc_id'][0:5]]
top5_sim_doc
```

```python
top5_sim_score=pd.concat([top5_sim_doc,sims_doc_sorted['sims'][0:5]],axis=1)
```
